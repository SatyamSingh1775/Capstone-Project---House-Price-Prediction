```python
sns.distplot(df['price'])
```

```python
sns.boxplot(x= df['price'])
```

```python
#Calculate the IQR of the price column
```

```python
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)

IQR = Q3 - Q1

#Define the bound for the outlier
lower_bound = Q1- 1.5*IQR
upper_bound = Q3 + 1.5*IQR

#Identify the outlier 

outliers = df[(df['price'] < lower_bound) | (df['price'] > upper_bound)]
```

```python
#Display the no of the outlier and the sam,e statistics 
```

```python
num_outliers = outliers.shape[0]
outlier_price_stats = outliers['price'].describe()

```

```python
num_outliers, outlier_price_stats
```

### price per sqrt

```python
sns.distplot(df['price_per_sqft'])
```

```python
sns.boxplot(x= df['price_per_sqft'])
```

```python
# Calculate the IQR of the price column 
```

```python
Q1 = df['price_per_sqft'].quantile(0.25)
Q3 = df['price_per_sqft'].quantile(0.75)

IQR = Q3 - Q1

#Define the bound of the outlier

lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR

#Identify the outlier

outlier_sqft = df[(df['price_per_sqft'] < lower_bound) | (df['price_per_sqft'] > upper_bound)]
```

```python
# Displaying the no of the outier and the some statistics 
```

```python
num_outlier = outlier_sqft.shape[0]

outlier_sqft_stats= outlier_sqft['price_per_sqft'].describe()
```

```python
num_outlier ,outlier_sqft_stats
```

```python
outlier_sqft['area'] = outlier_sqft['area'].apply(lambda x: x*9 if x<1000 else x)
```

```python
#This line calculates how much each square foot costs, using the price in rupees and area in sq ft.
```

```python

outlier_sqft['price_per_sqft'] = round((outlier_sqft['price']*10000000)/outlier_sqft['area'])
```
